```python
import matplotlib.pyplot as plt
import math
```

```python
#Exercise 1

def x_squared_minus_x_plus_two(x):
    y = (x**2)-x+2
    return y

x = range(0, 100)
y = [x_squared_minus_x_plus_two(n) for n in x]

plt.plot(x, y)
plt.annotate('Origin', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'pink'})
plt.show()
```

```python
#Exercise 2

#y=√x
x = range(0, 150)
y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]

plt.plot(x, y1, label='y=√x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Exercise 2')

plt.legend()

plt.show()

#y=x^3
plt.plot(x, y2, label='y=x^3')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Exercise 2')

plt.legend()

plt.show()

#y=2^x ver 1
plt.plot(x, y3, label='y=2^x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Exercise 2')

plt.legend()

plt.show()

#y=2^x ver 2
plt.plot(x, y3, linestyle='--', color='pink', linewidth='4', label='y=2^x ')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Exercise 2')

plt.legend()

plt.show()
```

```python
#Exercise 3

n_rows=1
n_cols=4

x = range(0, 150)
y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]

plt.figure(figsize=(14, 6))

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y1)
plt.title('y=√x')

plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y2)
plt.title('y=x^3')

plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y3)
plt.title('y=2^x')

plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y3, linestyle='--', color='pink', linewidth='4')
plt.title('y=2^x ')

plt.xlabel('x')
plt.ylabel('y')
plt.suptitle('Exercise 3')

plt.show()
```
