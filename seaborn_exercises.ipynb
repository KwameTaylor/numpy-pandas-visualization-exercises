```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
```

```python
# Exercise 0 - Iris database
```

```python
iris = sns.load_dataset('iris')
iris.to_csv('iris.csv')
iris.head()
```

```python
# What does the distribution of petal lengths look like?
# I am using a distribution chart.

sns.distplot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width?

sns.relplot(x="petal_length", y="petal_width", hue='species', data=iris)

print('We can see from the chart that there is a positive correlation between petal width and petal length.')
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length?
# I'm going to need to compare the sepal widths and lengths per species.

sns.relplot(x='sepal_width', y='sepal_length', hue='species', data=iris)

print('From the results of our scatter chart, we can see that it would be pretty reasonable to predict \nspecies based on sepal width and length for virginicas, but not as accurate for setosas or versicolors. Petal width is more accurate to determine between setosas and versicolors.')
```

```python
# Which features would be best used to predict species?

sns.pairplot(iris, hue='species')

print('Petal length and petal width are the best features to use in predicting species.')
```

```python
# Exercise 4 - chipotle

# Using the chipotle dataset from the previous exercise,
# create a bar chart that shows the 4 most popular items
# and the revenue produced by each.
```

```python
def get_db_url(db_name):
    from env import host, username, password
    url = f'mysql+pymysql://{username}:{password}@{host}/{db_name}'
    return (url)
```

```python
orders = pd.read_sql('''SELECT * FROM orders''',get_db_url('chipotle'))
orders.head()
```

```python
top_four = orders.groupby('item_name').quantity.sum().sort_values(ascending=False).head(4)
```

```python
# I'm a little confused about how to get the revenue in the chart as well as the quantity,
# as that's what gets me the most popular four items in the first place.

# Ask Samuel to see his code again
```

```python
top_four.plot(kind='barh',
             color='pink', 
             ec='black', 
             width=.8)

plt.title('The Big Four at Chipotle')
plt.ylabel('Menu Item')
plt.xlabel('Quanity Sold')

# reorder y-axis of horizontal bar chart
plt.gca().invert_yaxis()

plt.show()
```
