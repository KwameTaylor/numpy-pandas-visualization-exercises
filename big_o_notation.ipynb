```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#Exercise 5

x = range(1, 4)
y1 = [n*0+1 for n in x]
y2 = [np.log(n) for n in x]
y3 = [n for n in x]
y4 = [n*(np.log(n)) for n in x]
y5 = [n**2 for n in x]
y6 = [math.pow(2, n) for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.figure(figsize=(14, 6))

plt.plot(x, y1, label='O(1)')
plt.plot(x, y2, label='O(log n)')
plt.plot(x, y3, label='O(n)')
plt.plot(x, y4, label='O(n log n)')
plt.plot(x, y5, label='O(n^2)')
plt.plot(x, y6, label='O(2^n)')
plt.plot(x, y7, label='O(n!)')
plt.plot(x, y8, label='O(n^n)')


plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation')

plt.legend()

plt.show()
```
